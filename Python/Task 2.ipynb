```python
# Keep only needed columns
df_simple = df[["Product Name", "Price", "Discount", "Rating (detail)"]].copy()

print(df_simple.head())
```

```python
df_simple["Discount_Color"] = "None"
```

```python
df_simple.loc[df_simple["Discount"] > 50, "Discount_Color"] = "Green"
```

```python
df_simple.loc[df_simple["Discount"] < 10, "Discount_Color"] = "Red"
```

```python
print(df_simple[["Product Name", "Discount", "Discount_Color"]].head(10))
```

```python
#Create Rating_Color column
 #Default value “None”
df_simple["Rating_Color"] = "None"

# Set orange where Rating < 3
df_simple.loc[df_simple["Rating (detail)"] < 3, "Rating_Color"] = "Orange"
```

```python
print(df_simple[["Product Name", "Rating (detail)", "Rating_Color"]].head(10))
```
