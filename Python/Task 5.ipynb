```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# create dates from 1 to 7 November
dates = pd.date_range('2024-11-01', '2024-11-07')

# repeat these dates for all rows in df
df['scraping_date'] = [dates[i % len(dates)] for i in range(len(df))]
```

```python
print(df[['scraping_date']].head(15))
```

```python
daily_avg = df.groupby('scraping_date')['Discount'].mean().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot(daily_avg['scraping_date'], daily_avg['Discount'], marker='o')

plt.title('Average Discount Trend ')
plt.xlabel('Date')
plt.ylabel('Average Discount (%)')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
daily_avg['rolling_3d'] = daily_avg['Discount'].rolling(window=3, min_periods=1).mean()

plt.figure(figsize=(8,4))
plt.plot(daily_avg['scraping_date'], daily_avg['Discount'], marker='o', alpha=0.6, label='Daily avg')
plt.plot(daily_avg['scraping_date'], daily_avg['rolling_3d'], linewidth=2, label='3-day moving avg')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
